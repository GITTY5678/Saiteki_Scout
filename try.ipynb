```python
import ast

for name in dir(ast):
    obj = getattr(ast, name)

    if isinstance(obj, type) and issubclass(obj, ast.AST):
        print(name)
```

```python
import ast

code = """
n=int(input)
for i in range(n):
    print(i)
"""

tree = ast.parse(code)

print(ast.dump(tree, indent=4))
```

```python
import ast

code = """
n=int(input)
for i in range(n):
    for i in range(0,i):
        print(i)
"""

tree = ast.parse(code)

print(ast.dump(tree, indent=4))
```
